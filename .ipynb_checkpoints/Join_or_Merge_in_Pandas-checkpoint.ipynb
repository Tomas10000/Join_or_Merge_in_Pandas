```python
import pandas as pd
import numpy as np
```

```python
# data frame 1
d1 = {'Customer_id':pd.Series([1,2,3,4,5,6]),'Product':pd.Series(['Oven','Oven','Oven','Television','Television','Television'])}
df1 = pd.DataFrame(d1)
df1
```

```python
# data frame 2

d2 = {'Customer_id':pd.Series([2,4,6,7,8]),'State':pd.Series(['California','California','Texas','New York','Indiana'])}
df2 = pd.DataFrame(d2)
df2
```

```python
#inner join in python pandas
inner_join_df= pd.merge(df1, df2, on='Customer_id', how='inner')
inner_join_df 
```

```python
# outer join in python pandas
outer_join_df=pd.merge(df1, df2, on='Customer_id', how='outer')
outer_join_df
```

```python
# left join in python
left_join_df= pd.merge(df1, df2, on='Customer_id', how='left')
left_join_df 
```

```python
# right join in python pandas
right_join_df= pd.merge(df1, df2, on='Customer_id', how='right')
right_join_df 
```

**Simply concatenated both the tables based on their index**

```python
# join based on index python pandas

df_index = pd.merge(df1, df2, right_index=True, left_index=True)
df_index
```

**Concatenates two tables and keeps the old index**

```python
# Concatenate and keep the old index python pandas
df_row = pd.concat([df1, df2])
df_row
```

**Concatenate and change the index old**


```python
# Concatenate and change the index python pandas
df_row_reindex = pd.concat([df1, df2], ignore_index=True)
df_row_reindex
```

**Concatenate or join based on column index**

```python
# join based on index python pandas
df_col = pd.concat([df1,df2], axis=1)
df_col
```
